```python
# Select country France to do the Further analysis
# 
country_select="France"
country_select="Germany"

data=df[df.Country==country_select].groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
```

```python
data.shape
print("Number of Orders from",country_select,":",data.shape[0])
```

```python
data=data.unstack().reset_index().fillna('0')
```

```python
# data
```

```python
data.set_index('InvoiceNo').head()
```

```python
data.columns=data.columns.str.replace(' ','_')
```

```python

for cl in data.columns:
    data[cl]=data[cl].astype(int)
```

```python
def onehot(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
dataset=data.applymap(onehot)
dataset=dataset.set_index('InvoiceNo')
```

```python
frequent_itemsets= apriori(dataset, min_support=0.1, use_colnames=True)
frequent_itemsets['length Of Frequent Set'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.5)
```
